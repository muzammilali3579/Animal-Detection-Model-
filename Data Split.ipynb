```python
import os
import random
import shutil

dataset_dir = 'MY OWN DATASET/'
split_dir = 'Splitted Dataset/'

# Set the split ratios (adjust as needed)
train_ratio = 0.8
test_ratio = 0.1
valid_ratio = 0.1

# Create the destination directories for the split data
train_dir = os.path.join(split_dir, 'train')
test_dir = os.path.join(split_dir, 'test')
valid_dir = os.path.join(split_dir, 'validation')

os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
os.makedirs(valid_dir, exist_ok=True)

# Iterate over the subfolders in the dataset directory (each subfolder represents a class)
for class_name in os.listdir(dataset_dir):
    class_dir = os.path.join(dataset_dir, class_name)

    # Skip if the current item in the dataset directory is not a subfolder
    if not os.path.isdir(class_dir):
        continue

    # Create the destination subdirectories for the split data
    class_train_dir = os.path.join(train_dir, class_name)
    class_test_dir = os.path.join(test_dir, class_name)
    class_valid_dir = os.path.join(valid_dir, class_name)

    os.makedirs(class_train_dir, exist_ok=True)
    os.makedirs(class_test_dir, exist_ok=True)
    os.makedirs(class_valid_dir, exist_ok=True)

    # Get the list of image files in the current class directory
    image_files = os.listdir(class_dir)

    # Shuffle the image files randomly
    random.shuffle(image_files)

    # Split the image files into training, testing, and validation sets
    train_count = int(train_ratio * len(image_files))
    test_count = int(test_ratio * len(image_files))
    valid_count = len(image_files) - train_count - test_count

    train_files = image_files[:train_count]
    test_files = image_files[train_count:train_count + test_count]
    valid_files = image_files[train_count + test_count:]

    # Move the image files to the respective split directories
    for file in train_files:
        src_path = os.path.join(class_dir, file)
        dst_path = os.path.join(class_train_dir, file)
        shutil.copy(src_path, dst_path)

    for file in test_files:
        src_path = os.path.join(class_dir, file)
        dst_path = os.path.join(class_test_dir, file)
        shutil.copy(src_path, dst_path)

    for file in valid_files:
        src_path = os.path.join(class_dir, file)
        dst_path = os.path.join(class_valid_dir, file)
        shutil.copy(src_path, dst_path)

    print(f'Split class "{class_name}": Train: {len(train_files)}, Test: {len(test_files)}, Validation: {len(valid_files)}')
```
