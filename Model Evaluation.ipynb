```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Load the trained model
model = load_model('animal_recognition_model.h5')  # Replace with the path to your trained model

# Set up the data directories
test_dir = 'Dataset/test'  # Replace with the path to your test data directory
validation_dir = 'Dataset/validation'  # Replace with the path to your validation data directory

# Set up the hyperparameters
batch_size = 32

# Create data generators for test and validation data
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

validation_datagen = ImageDataGenerator(rescale=1./255)
validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Evaluate the model on the test data
test_loss, test_accuracy = model.evaluate(test_generator)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)

# Make predictions on the validation data
validation_predictions = model.predict(validation_generator)

# Get the predicted class labels for validation data
predicted_labels = validation_predictions.argmax(axis=1)
class_labels = list(validation_generator.class_indices.keys())

# Print the predicted class labels for validation data
for i, label in enumerate(predicted_labels):
    print("Image", i+1, "predicted as", class_labels[label])
```
